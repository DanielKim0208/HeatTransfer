```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Raw data plot
df1 = pd.read_excel('HeatTransfer_HW1_plots.xlsx', sheet_name='RawData')
display(df1)
plt.scatter(df1["time (min)"], df1["T1, wooden floor"], label="T1, wooden floor", marker="x")
plt.scatter(df1["time (min)"], df1["T2_MiracleThaw"], label="T2, MiracleThaw")
plt.xlabel("Time (min)")
plt.ylabel("Temperature(C)")
plt.legend()
plt.title("Temperature of different surfaces vs. Time")
plt.show()
```

```python
#Dimensionless temperature and time constant 
df2 = pd.read_excel('HeatTransfer_HW1_plots.xlsx', sheet_name='Dimensionless Temperature')
display(df2)
plt.scatter(df2["time (min)"], df2["q1"], label="wooden floor", marker="x")
slope_wooden_floor, intercept_wooden_floor = np.polyfit(df2["time (min)"], df2["q1"], 1)
plt.plot(df2["time (min)"], slope_wooden_floor * df2["time (min)"] + intercept_wooden_floor, color='blue', linestyle='--', label=f'Linear Fit (wooden floor): t/tau = {slope_wooden_floor:.4f}')
plt.scatter(df1["time (min)"], df2["q2"], label="MiracleThaw")
slope_miracle_floor, intercept_miracle_floor = np.polyfit(df2["time (min)"], df2["q2"], 1)
plt.plot(df2["time (min)"], slope_miracle_floor * df2["time (min)"] + intercept_miracle_floor, color='red', linestyle='dotted', label=f'Linear Fit (MiracleThaw): t/tau = {slope_miracle_floor:.4f}')




plt.xlabel("Time (min)")
plt.ylabel("Dimensionless Temperature")
plt.yscale("log")
plt.legend(loc = "best")
plt.title("Dimensionless Temperature of different surfaces vs. Time")
plt.show()
```
